```python
import pandas as pd
df = pd.read_csv('Car_sales - Car_sales.csv')
df.head()
df.describe()
df.info()
```

```python
#Identify the variable types in the given data set
df.info()
```

```python
missing_count = df.isnull().sum()
print("\nMissing Values Count:",missing_count)
```

```python
duplicate_rows = df.duplicated()

# Print the original dataframe and the duplicate rows
print("Original DataFrame:")
print(df)
print("\nDuplicate Rows:")
print(duplicate_rows)
```

```python
df_no_duplicates = df.drop_duplicates()
print("DataFrame without Duplicates:")
print(df_no_duplicates)
```

```python
print("\nFirst 5 rows of the DataFrame:")
df.tail(15)
```

```python
#Make a pie chart for showing vehicle type by percentage
import pandas as pd
import matplotlib.pyplot as plt
vehicle_counts = df['Vehicle_type'].value_counts()
Vehicle_type_percentage = (vehicle_counts / len(df)) * 100
Vehicle_type_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title('Vehicle_type ')
plt.ylabel('') 
plt.show()
```

```python
#Make a bar graph for manufacturer type
plt.figure(figsize=(10, 8))
manufacturer_counts = df['Manufacturer'].value_counts()
plt.bar(manufacturer_counts.index, manufacturer_counts, color='blue')
plt.xlabel('Manufacturer')
plt.ylabel('Manufacturer Count')
plt.title('Manufacturer Types Bar Graph')
plt.show()
```

```python
#Calculate average Price of all cars by using correct measure of central Tendency
car_price = df[df['Vehicle_type']=='Car']['Price_in_thousands']
average_car_price = car_price.mean()
print("The average price of cars is:",average_car_price)
```

```python
#Calculate the variance and standard deviation for numerical columns such as
#Sales_in_thousands, __year_resale_value, Price_in_thousands, Engine_size, Horsepower

sales_variance_population = df['Sales_in_thousands'].var(ddof=0)
sales_std_dev_population = df['Sales_in_thousands'].std(ddof=0)
print("Population Sales Variance:", sales_variance_population)
print("Population Sales Standard Deviation:", sales_std_dev_population)
```

```python
resale_variance = df['__year_resale_value'].var(ddof=0)
resale_std_dev = df['__year_resale_value'].std(ddof=0)
print("Resale Value Variance:", resale_variance)
print("Resale Value Standard Deviation:", resale_std_dev)
```

```python
price_variance = df['Price_in_thousands'].var(ddof=0)
price_std_dev = df['Price_in_thousands'].std(ddof=0)
print("Price Variance:", price_variance)
print("Price Standard Deviation:", price_std_dev)
```

```python
Engine_size_variance = df['Engine_size'].var(ddof=0)
Engine_size_std_dev = df['Engine_size'].std(ddof=0)
print("Engine size Variance:", Engine_size_variance)
print("Engine size Standard Deviation:", Engine_size_std_dev)
```

```python
horsepower_variance = df['Horsepower'].var(ddof=0)
horsepower_std_dev = df['Horsepower'].std(ddof=0)
print("Horsepower Variance:", horsepower_variance)
print("owrsepower Sales Standard Deviation:", horsepower_std_dev)
```

```python
#For the same columns Calculate IQR and identify outliers in these columns
Q1 = df['Sales_in_thousands'].quantile(0.25)
Q3 = df['Sales_in_thousands'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Sales_in_thousands'] < Q1 - 1.5 * IQR) | (df['Sales_in_thousands']> Q3 + 1.5 * IQR)]
print("Q1=",Q1)
print("Q3=",Q3)
print("IQR=",IQR)
print("Outliers",outliers)
```
